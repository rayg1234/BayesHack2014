```python
names = pd.read_csv('names.txt', sep = '\t')
names = names[names.Name.apply(lambda x: len(str(x)) >= 3)]
names['Name'] = names.Name.apply(lambda x : str(x).lower())
names =  names[names.Rank < 1000]
fem_names = list(names.Name)
print fem_names
```

```python
def process_names(x):
    try:
        x = x.lower()
        found = []
        for name in fem_names:
            search = ' {} '.format(name)
            if x.find(search) > -1:
                found.append(name)
    except Exception as e:
        print e
        found = []
    return found
```

```python
def get_age(x):
    try:
        return int(float(x))
    except Exception:
        return -1

def phone_agg(gr):
    ages = list(gr.postAge.apply(get_age))
    unique_ages = list(np.unique(np.array(ages)))
    if -1 in unique_ages:
        unique_ages.remove(-1)
    num_unique_ages = len(unique_ages)
    num_same_phone = gr.postPhone.count()
    names = list(gr.postText.apply(process_names))
    names = list(set(x[0] for x in names if len(x)))
    num_names = len(names)
    cities = np.array(gr.City.unique())
    num_cities = len(cities)
    return pd.Series({'num_same_phone':num_same_phone,
                      'num_unique_ages':num_unique_ages,
                      'unique_ages':unique_ages,
                      'num_names':num_names,
                      'names':names,
                      'cities':cities,
                      'num_cities':num_cities
                      })

grouped_sample = df_sample.groupby('postPhone').apply(phone_agg)
# grouped = df.groupby('postPhone').apply(phone_agg)
```

```python
import re

def get_email(x):
    try:
        text = x.lower()
        email = re.findall('[\w]+@[a-z0-9]+\.com', text)
        if len(email):
            return email[0]
    except Exception as e:
        pass
    return np.nan

df_sample['email'] = df_sample.postText.apply(get_email)
```
